```python
dfTmp['continent'] = dfTmp['country'].map(paisAcontinente)
dfTmp.head()
```

```python
def get_continentID(dfTmp):
    le = LabelEncoder()
    le.fit(dfTmp['continent'])
    return le.transform(dfTmp['continent'])
```

```python
#le = LabelEncoder()
#le.fit(dfTmp['continent'])
#le.transform(dfTmp['continent'])
```

```python
#dfTmp['continentId'] = le.transform(dfTmp['continent'])
dfTmp['continentId'] = get_continentID(dfTmp)
dfTmp.head()
```

```python
dfTmp['is_sweet'] = dfTmp.description.str.contains('sweet', regex=False)
dfTmp.head()
```

```python
bins = [0,15,40,70,3400]
dfTmp['price_4levels'] = pd.cut(df.price, bins, labels =[0, 1, 2, 3])
dfTmp.head()
```

```python
dfTmp['was_master_taster'] = dfTmp.taster_name.str.contains('Roger Voss|Michael Schachner|Paul Gregutt|Virginie Boone', regex=True)
dfTmp.head()
```

```python
bins = [1900,1999,2019]
dfTmp['century_title'] = pd.cut(pd.to_numeric(df.title.str.extract(r'.*(\d{4}).*', expand=False), errors='coerce'), bins, labels =[0, 1])
dfTmp.head()
```

```python
dfTmp.dropna(subset=['price'], inplace=True)
dfTmp.dropna(subset=['century_title'], inplace=True)
```

```python
dfTmp.head()
```

```python
dfTmp = dfTmp.sample(frac=1).reset_index(drop=True)
```

```python
dfTmp.head()
```

```python
msk = np.random.rand(len(dfTmp)) < 0.8
train = dfTmp[msk]
test = dfTmp[~msk]
print(dfTmp.shape[0])
print(train.shape[0])
print(test.shape[0])
```

```python
X = train[['price']].values
Y = train['points'].values
reg = LinearRegression().fit(X, Y)
reg.score(X, Y)
```

```python
y_true = test['points']
y_pred = reg.predict(test[['price']])
r2_score(y_true, y_pred)
```

```python
X = train[['price', 'century_title']].values
Y = train['points'].values
reg = LinearRegression().fit(X, Y)
reg.score(X, Y)
```

```python
y_true = test['points']
y_pred = reg.predict(test[['price', 'century_title']])
r2_score(y_true, y_pred)
```

```python
X = train[['price', 'century_title', 'continentId']].values
Y = train['points'].values
reg = LinearRegression().fit(X, Y)
reg.score(X, Y)
```

```python
y_true = test['points']
y_pred = reg.predict(test[['price', 'century_title','continentId']])
r2_score(y_true, y_pred)
```
